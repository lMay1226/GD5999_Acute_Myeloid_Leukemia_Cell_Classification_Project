```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, cohen_kappa_score

# Load the CSV file from the given path
file_path = '/home/ym62/group_re_annotations.csv'
data = pd.read_csv(file_path)

# Filter out rows where the second column is 'UNC'
filtered_data = data[data['first re-annotation'] != 'UNC']

# Get the true and predicted values
y_true = filtered_data['cell types']
y_pred = filtered_data['first re-annotation']

# Compute the confusion matrix
cm = confusion_matrix(y_true, y_pred, labels=y_true.unique())

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=y_true.unique(), yticklabels=y_true.unique())
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix between First Annotator and Re-Annotator')
#plt.xticks(rotation=45)
#plt.yticks(rotation=45)
plt.tight_layout()
plt.show()

# Calculate Cohen's Kappa score
kappa_score = cohen_kappa_score(y_true, y_pred)

print(f"Cohen's Kappa Score (Excluding UNC in Second Column): {kappa_score:.3f}")
```

```python
# Filter out rows where the third column is 'UNC'
filtered_data = data[data.iloc[:, 2] != 'UNC']

# Extract the first and third columns
gold_standard = filtered_data.iloc[:, 0]
second_re_annotation = filtered_data.iloc[:, 2]

# Create the confusion matrix
conf_matrix = confusion_matrix(gold_standard, second_re_annotation)

# Get the unique class names
class_names = sorted(list(set(gold_standard) | set(second_re_annotation)))

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix: Gold Standard vs Second Re-Annotation")
plt.xlabel("Second Re-Annotation")
plt.ylabel("Gold Standard")
#plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Calculate Cohen's Kappa score
kappa_score = cohen_kappa_score(gold_standard, second_re_annotation)

print(f"Cohen's Kappa Score (Excluding UNC in Second Column): {kappa_score:.3f}")
```
